```python
import os
from glob import glob
from sklearn.model_selection import train_test_split
import shutil


img_files = sorted(glob('../NIH-CXR-MSUFAL/Photographs/*.*'))
annot_files = sorted(glob('../NIH-CXR-MSUFAL/Annotations/*.*'))


# Define the directories for train, val, and test data
train_dir = 'train'
val_dir = 'val'
test_dir = 'test'

# Split the data into train, val, and test sets
img_train, img_valtest, annot_train, annot_valtest = train_test_split(img_files, annot_files, test_size=0.2, random_state=42)
img_val, img_test, annot_val, annot_test = train_test_split(img_valtest, annot_valtest, test_size=0.5, random_state=42)

# Create the train, val, and test directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Move the image files to the train directory
for img_file in img_train:
    shutil.move(img_file, os.path.join(train_dir, os.path.basename(img_file)))

# Move the annotation files to the train directory
for annot_file in annot_train:
    shutil.move(annot_file, os.path.join(train_dir, os.path.basename(annot_file)))

# Move the image files to the val directory
for img_file in img_val:
    shutil.move(img_file, os.path.join(val_dir, os.path.basename(img_file)))

# Move the annotation files to the val directory
for annot_file in annot_val:
    shutil.move(annot_file, os.path.join(val_dir, os.path.basename(annot_file)))

# Move the image files to the test directory
for img_file in img_test:
    shutil.move(img_file, os.path.join(test_dir, os.path.basename(img_file)))

# Move the annotation files to the test directory
for annot_file in annot_test:
    shutil.move(annot_file, os.path.join(test_dir, os.path.basename(annot_file)))
# Split into train valid and test sets at ratio 80:10:10 randomly

  
```

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import matplotlib

matplotlib.use('TkAgg')

model = YOLO("/home/sonymd/Desktop/YOLO8/runs/detect/train/weights/best.pt")


out = model(['/home/sonymd/Desktop/YOLO8/NIH-CXR-MSUFAL/test/images/1305_AM_Chest_AP8.png',
             '/home/sonymd/Desktop/YOLO8/NIH-CXR-MSUFAL/test/images/1296_AM_Chest_AP24.png',
             '/home/sonymd/Desktop/YOLO8/NIH-CXR-MSUFAL/test/images/1354_PM_Chest_AP1.png'])
len(out)
out = out[1]


predicted_classes = out.boxes.cls.cpu().numpy().tolist()
predicted_boxes = out.boxes.xyxy.cpu().numpy()
predicted_scores = out.boxes.conf.cpu().numpy()
class_names = out.names
original_image = out.orig_img


# Extract the bounding box coordinates and class labels

def load_annotations(annot_path):
    with open(annot_path, 'r') as f:
        lines = f.readlines()
        boxes = [list(map(float, line.strip().split())) for line in lines]
    return boxes


def load_gt_box(boxes, width, height, class_id):
    # extract the box with class label class
    idx = [box[0] for box in boxes].index(class_id)

    # convert yolo format to x1, y1, x2, y2
    xmin = int((boxes[idx][1] - boxes[idx][3] / 2) * width)
    ymin = int((boxes[idx][2] - boxes[idx][4] / 2) * height)
    xmax = int((boxes[idx][1] + boxes[idx][3] / 2) * width)
    ymax = int((boxes[idx][2] + boxes[idx][4] / 2) * height)

    return xmin, ymin, xmax, ymax


img_path = out.path
annot_path = img_path.replace('images', 'labels').replace('.png', '.txt')
boxes = load_annotations(annot_path)
width, height = out.orig_img.shape[1], out.orig_img.shape[0]


plt.figure(figsize=(10, 10), dpi=100)

plt.subplot(2, 2, 1)
plt.imshow(original_image)


# Show the predicted bounding box T1-T5
plt.subplot(2, 2, 2)
# plot the ground truth box
xmin, ymin, xmax, ymax = load_gt_box(boxes, width, height, 0)
img = cv2.rectangle(original_image.copy(), (xmin, ymin),
                    (xmax, ymax), (0, 255, 0), 4)

# Plot the predicted box
idx = predicted_classes.index(0)
xmin, ymin, xmax, ymax = map(int, predicted_boxes[idx])
img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (255, 0, 0), 4)
plt.imshow(img)
plt.title(out.names[0])


# Show the Clavicle Only
plt.subplot(2, 2, 3)
# plot the ground truth box
xmin, ymin, xmax, ymax = load_gt_box(boxes, width, height, 1)
img = cv2.rectangle(original_image.copy(), (xmin, ymin),
                    (xmax, ymax), (0, 255, 0), 4)

# Plot the predicted box
idx = predicted_classes.index(1)
xmin, ymin, xmax, ymax = map(int, predicted_boxes[idx])
img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (255, 0, 0), 4)
plt.imshow(img)
plt.title(out.names[1])


# Show the Complete Vertebrae
plt.subplot(2, 2, 4)
# plot the ground truth box
xmin, ymin, xmax, ymax = load_gt_box(boxes, width, height, 2)
img = cv2.rectangle(original_image.copy(), (xmin, ymin),
                    (xmax, ymax), (0, 255, 0), 4)

# Plot the predicted box
idx = predicted_classes.index(2)
xmin, ymin, xmax, ymax = map(int, predicted_boxes[idx])
img = cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (255, 0, 0), 4)
plt.imshow(img)
plt.title(out.names[2])
plt.show()
```
